```python
with open('data/salmonella_spi1_region.fna', 'r') as f:
    salmon_seq = ''
    for line in f:
        if line[:3] != '>gi':
            line = f.readline().rstrip()
            salmon_seq += line
    print(salmon_seq[:1000])
```

```python
def gc_blocks(seq, block_size):
    """Divides a sequence into blocks and computes the GC content for each block, returning a tuple."""
    
    i = 0
    gc_list = []
    while i < len(seq) - block_size + 1:
        block = seq[i:i + block_size]
        gc_counter = block.count('G') + block.count('C')
        print(block)
        print(gc_counter / block_size)
        gc_content = gc_counter / block_size
        gc_list += [gc_content]
        i += block_size
    gc_tuple = tuple(gc_list)
    return gc_tuple
    
```

```python
seq = salmon_seq[:1000]

gc_blocks(seq, 100)
```

```python
def gc_map(seq, block_size, gc_thresh):
    """Divides a sequence into blocks, computes the GC content for each block, 
    and returns a gc map."""
    
    i = 0
    gc_list = []
    gc_string = ''
    while i < len(seq) - block_size + 1:
        block = seq[i:i + block_size]
        gc_counter = block.count('G') + block.count('C')
        gc_content = gc_counter / block_size
        gc_list += [gc_content]
        i += block_size
        if gc_content > gc_thresh:
            gc_string += block.upper()
        else:
            gc_string += block.lower()
    return gc_string
    
```

```python
seq = salmon_seq

gc_map(seq, 50, .45)
```
